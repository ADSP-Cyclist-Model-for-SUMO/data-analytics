```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%store -r durations_b1
%store -r durations_chb
%store -r durations_moabit
%store -r durations_oranienstr
%store -r durations_prinzenstr

duration_data = [durations_oranienstr, durations_moabit, durations_chb, durations_b1 , durations_prinzenstr ]
duration_data_subset = [durations_oranienstr, durations_b1]
```

```python
defaults = []
defaults_ped = []
new_params = []
new_params_ped = []
simra = []

for arr in duration_data:
    defaults.append(arr[0])
    defaults_ped.append(arr[1])
    new_params.append(arr[2])
    new_params_ped.append(arr[3])
    simra.append(arr[4])
```

```python
from matplotlib import rc

ticks = ['Oranienstr., Kreuzberg', 'Waldstraße, Moabit', 'Fasanenstr., CHB', 'Frankfurter Allee, Friedrichsh.', 'Prinzenstr., Kreuzberg']

def set_box_color(bp, color):
    plt.setp(bp['boxes'], color=color)
    plt.setp(bp['whiskers'], color=color)
    plt.setp(bp['caps'], color=color)
    plt.setp(bp['medians'], color=color)

ratio = 6
fig_kw = dict(figsize=(16,20))
f, (ax, ax2) = plt.subplots(2, 1, sharex=True, **fig_kw, gridspec_kw={'height_ratios': [1,ratio]})
rc('font', **{'family': 'serif', 'serif': ['Computer Modern'], 'size': 16})
rc('text', usetex=True)
bpd = ax.boxplot(defaults, positions=np.array(range(len(defaults)))*5.0-1.6, widths=0.6, sym='')
bpdp = ax.boxplot(defaults_ped, positions=np.array(range(len(defaults)))*5.0-0.8, widths=0.6, sym='')
bpn = ax.boxplot(new_params, positions=np.array(range(len(new_params)))*5.0, widths=0.6, sym='')
bpnp = ax.boxplot(new_params_ped, positions=np.array(range(len(new_params)))*5.0 + 0.8, widths=0.6, sym='')
bps = ax.boxplot(simra, positions=np.array(range(len(simra)))*5.0+1.6, widths=0.6, sym='')


bps2 = ax2.boxplot(simra, positions=np.array(range(len(simra)))*5.0+1.6, widths=0.6, sym='')
bpd2 = ax2.boxplot(defaults, positions=np.array(range(len(defaults)))*5.0-1.6, widths=0.6, sym='')
bpdp2 = ax2.boxplot(defaults_ped, positions=np.array(range(len(defaults)))*5.0-0.8, widths=0.6, sym='')
bpn2 = ax2.boxplot(new_params, positions=np.array(range(len(new_params)))*5.0, widths=0.6, sym='')
bpnp2 = ax2.boxplot(new_params_ped, positions=np.array(range(len(new_params)))*5.0 + 0.8, widths=0.6, sym='')
bps2 = ax2.boxplot(simra, positions=np.array(range(len(simra)))*5.0+1.6, widths=0.6, sym='')

colors = [u'#1f77b4', u'#ff7f0e', u'#2ca02c']

set_box_color(bpd, 'r')
set_box_color(bpdp, '#c51b8a')
set_box_color(bpn, u'#ff7f0e')
set_box_color(bpnp, 'g')
set_box_color(bps, u'#1f77b4')

set_box_color(bpd2, 'r')
set_box_color(bpdp2, '#c51b8a')
set_box_color(bpn2, u'#ff7f0e')
set_box_color(bpnp2, 'g')
set_box_color(bps2, u'#1f77b4')

# draw temporary lines and use them to create a legend
ax.plot([], c='r', label='Slow Cars SUMO Default')
ax.plot([], c='#c51b8a', label='Fast Pedestrians SUMO Default')
ax.plot([], c=u'#ff7f0e', label='Slow Cars SUMO Parametrized')
ax.plot([], c='g', label='Fast Pedestrians SUMO Parametrized')
ax.plot([], c=u'#1f77b4', label='SimRa')

ax.set_ylim(1150,1950)
ax2.set_ylim(0,900)

ax.spines['bottom'].set_visible(False)

ax2.spines['top'].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop=False) 
ax2.xaxis.tick_bottom()

d = .005
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d*ratio, +d*ratio), **kwargs)
ax.plot((1 - d, 1 + d), (-d*ratio, +d*ratio), **kwargs)

kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)

ax.legend(fontsize=16)
ax.set_yticks([1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900])
ax.grid(axis='y', linewidth=0.5)
ax2.grid(axis='y', linewidth=0.5)

plt.xlabel('Scenario')
plt.ylabel('Duration in seconds')

[ax.axvline(x,color= '#bdbdbd', linewidth=0.5, linestyle='-') for x in [2.5, 7.5, 12.5, 17.5]] 
[ax2.axvline(x,color= '#bdbdbd', linewidth=0.5, linestyle='-') for x in [2.5, 7.5, 12.5, 17.5]] 

plt.xticks(range(0, len(ticks) * 5, 5), ticks)
f.tight_layout()

plt.savefig('images/sim_complete_full.png', dpi=300, bbox_inches='tight')
```
